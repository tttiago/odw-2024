```python
# Download the challenge set files
from pycbc.frame import read_frame
import urllib.request

fname = 'challenge1.gwf'
url = f"https://dcc.ligo.org/public/0187/G2300818/001/{fname}"
urllib.request.urlretrieve(url, fname)
print(f'Getting : {url}')

# LOSC bulk data typically uses the same convention for internal channels names
# Strain is typically IFO:LOSC-STRAIN, where IFO can be H1/L1/V1.
channel_name = "H1:CHALLENGE1"

ts = read_frame(fname, channel_name)
```

```python
print(f"Duration: {ts.duration} s")
print(f"Sample rate: {ts.sample_rate} Hz")
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-poster")
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(ts.sample_times.data, ts.data, linewidth=1, label='H1')
plt.xlabel("Time (s)")
plt.ylabel("Strain")
plt.xlim([0, 128])
plt.legend()
```
